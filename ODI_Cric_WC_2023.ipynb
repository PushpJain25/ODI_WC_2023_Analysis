COLLECTING SCORECARD LINKS FROM CRICBUZZ.com

```python
headers1 = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
webpage1 = requests.get('https://www.cricbuzz.com/cricket-series/6732/icc-cricket-world-cup-2023/matches',headers=headers1).text
```

```python
soup = BeautifulSoup(webpage1, 'lxml')
print(soup.prettify())
```

```python
#Collecting full details of each match , i.e. Collecting particular match links and joining the base url with href tags
match_links=[]
base_url = "https://www.cricbuzz.com/"
for i in soup.find_all('a',class_="text-hvr-underline",href=True):
    relative_url = i['href']
    absolute_url = urljoin(base_url, relative_url)
    match_links.append(absolute_url)
del match_links[49]
del match_links[48]
match_links
```

```python
base_url = "https://www.cricbuzz.com/"

for a_tag in soup.find_all('a',class_="cb-nav-tab", href=True):
    relative_url = a_tag['href']
    absolute_url = urljoin(base_url, relative_url)
    print(absolute_url)
```

```python
webpage_links = match_links
specific_links = []
def extract_specific_link(url):
    response = requests.get(url)
    
    if response.status_code == 200:
        soup1 = BeautifulSoup(response.content, 'html.parser')
        specific_link = soup1.find_all('a',class_="cb-nav-tab",href=True)
        
        if specific_link:
            specific_links.append(specific_link)

for link in webpage_links:
    extract_specific_link(link)
```

```python
#All the links of tabs on the webpage of each match have been stored collectively , and its list of lists
specific_links
```

```python
#Extracting only scorecard link out of all other tab links from all the sublist in specific_links list
href_values = []

for sublist in specific_links:
    if len(sublist) > 1:
        href_value = sublist[1].get('href')
        if href_value:
            absolute_url = urljoin(base_url, href_value)
            href_values.append(absolute_url)

print(href_values)
```

```python
team_1_score = []
team_2_score = []
match_nmber=[]
def teams_scores(urls):
    match_id=0
    for url in urls:
        response = requests.get(url, headers=headers1)
        if response.status_code == 200:
            # Parse the webpage content using BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')

            for i in soup.find_all('div', id='innings_1'):
                score = i.find('span', class_='pull-right').text
                score = score.split(" ")
                team_1_score.append(score[0])

            for i in soup.find_all('div', id='innings_2'):
                score = i.find('span', class_='pull-right').text
                score = score.split(" ")
                team_2_score.append(score[0])
            match_id+=1
            match_nmber.append(match_id)
```

WRITING A FUNCTION WHICH WILL PARSE EACH SCORECARD LINK COLLECT ALL DETAILS OF THAT PARTICULAR MATCH

FIRSTLY , BATTING DETAILS

```python
balls = []
batsman = []
dismissal = []
runs = []
sixes = []
fours = []
strike_rate = []
match = []
team_innings = []
batsman_pos=[]
match_no=[]
```

```python
def batting_data(urls):
    match_id=0
    for url in urls:
        response = requests.get(url, headers=headers1)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            
            #1st Batting Innings
            s = soup.find('h1').text
            new_s = s.split(',')
            for i in soup.find_all('div', id='innings_1'):
                t = i.find('span').text
                
            match_id+=1    
            pos=0
            for i in soup.find_all('div', id='innings_1'):
                for j in i.find_all("div", class_="cb-col cb-col-25"):
                    pos=pos+1
                    batsman_pos.append(pos)
                    batsman.append(j.text)
                    match.append(new_s[0])
                    team_innings.append(t)
                    match_no.append(match_id)

            for i in soup.find_all('div', id='innings_1'):
                for j in i.find_all("div", class_="cb-col cb-col-33"):
                    for k in j.find_all('span', class_='text-gray'):
                        dismissal.append(k.text)

            first_div_tag = soup.find("div", class_="cb-col cb-col-100 cb-ltst-wgt-hdr")
            for k in first_div_tag.find_all('div', class_="cb-col cb-col-8 text-right text-bold"):
                runs.append(k.text)
            runs.remove('R')

            temp = []
            first_div_tag = soup.find("div", class_="cb-col cb-col-100 cb-ltst-wgt-hdr")
            for k in first_div_tag.find_all('div', class_="cb-col cb-col-8 text-right"):
                temp.append(k.text)

            del temp[0:4]
            for l in range(0, len(temp), 4):
                balls.append(temp[l])
                fours.append(temp[l + 1])
                sixes.append(temp[l + 2])
                strike_rate.append(temp[l + 3])
                
            #2nd Batting Innings
            for i in soup.find_all('div', id='innings_2'):
                t = i.find('span').text
                
            pos=0
            for i in soup.find_all('div', id='innings_2'):
                for j in i.find_all("div", class_="cb-col cb-col-25"):
                    pos=pos+1
                    batsman_pos.append(pos)
                    batsman.append(j.text)
                    match.append(new_s[0])
                    team_innings.append(t)
                    match_no.append(match_id)

            for i in soup.find_all('div', id='innings_2'):
                for j in i.find_all("div", class_="cb-col cb-col-33"):
                    for k in j.find_all('span', class_='text-gray'):
                        dismissal.append(k.text)

            second_parent = soup.find('div', id="innings_2")
            div_element = second_parent.find('div', class_='cb-col cb-col-100 cb-ltst-wgt-hdr')
            for i in div_element.find_all('div', class_="cb-col cb-col-8 text-right text-bold"):
                runs.append(i.text)
            runs.remove('R')

            temp = []
            second_parent = soup.find('div', id="innings_2")
            div_element = second_parent.find('div', class_='cb-col cb-col-100 cb-ltst-wgt-hdr')

            for k in div_element.find_all('div', class_="cb-col cb-col-8 text-right"):
                temp.append(k.text)

            del temp[0:4]
            for l in range(0, len(temp), 4):
                balls.append(temp[l])
                fours.append(temp[l + 1])
                sixes.append(temp[l + 2])
                strike_rate.append(temp[l + 3])
```

```python
#Passing all the scorecard links to function batting_data
batting_data(href_values)
```

NOW , BOWLING DETAILS

```python
bowlers = []
overs = []
wickets = []
maidens = []
wide = []
no_balls = []
economy = []
given_runs = []
match_name = []
innings = []
match_no2=[]
```

```python
def bowling_data(urls):
    match_id=0
    for url in urls:
        response = requests.get(url, headers=headers1)
        if response.status_code == 200:
            # Parse the webpage content using BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')
            
            #1st Bowling Details
            sub_divs = soup.find_all('div', class_='cb-col cb-col-100 cb-ltst-wgt-hdr')
            bowlers_table = sub_divs[1]
            
            match_id+=1
            s = soup.find('h1').text
            new_s = s.split(',')
            
            for i in soup.find_all('div', id='innings_1'):
                t1 = i.find('span').text
            for i in soup.find_all('div', id='innings_2'):
                t2 = i.find('span').text
            for i in bowlers_table.find_all('div', class_="cb-col cb-col-38"):
                bowlers.append(i.text)
                match_name.append(new_s[0])
                innings.append(t2)
                match_no2.append(match_id)

            for i in bowlers_table.find_all('div', class_="cb-col cb-col-8 text-right text-bold"):
                wickets.append(i.text)

            temp = []
            for i in bowlers_table.find_all('div', class_="cb-col cb-col-8 text-right"):
                temp.append(i.text)
            del temp[0:4]
            for l in range(0, len(temp), 4):
                overs.append(temp[l])
                maidens.append(temp[l + 1])
                no_balls.append(temp[l + 2])
                wide.append(temp[l + 3])

            temp = []
            for i in bowlers_table.find_all('div', class_="cb-col cb-col-10 text-right"):
                temp.append(i.text)
            del temp[0:2]
            for l in range(0, len(temp), 2):
                given_runs.append(temp[l])
                economy.append(temp[l + 1])
                
            #2nd Bowling Innings
            divs = soup.find('div', id='innings_2')
            sub_divs = divs.find_all('div', class_='cb-col cb-col-100 cb-ltst-wgt-hdr')
            bowlers_table_2 = sub_divs[1]

            for i in bowlers_table_2.find_all('div', class_="cb-col cb-col-38"):
                bowlers.append(i.text)
                match_name.append(new_s[0])
                innings.append(t1)
                match_no2.append(match_id)

            for i in bowlers_table_2.find_all('div', class_="cb-col cb-col-8 text-right text-bold"):
                wickets.append(i.text)

            temp = []
            for i in bowlers_table_2.find_all('div', class_="cb-col cb-col-8 text-right"):
                temp.append(i.text)
            del temp[0:4]
            for l in range(0, len(temp), 4):
                overs.append(temp[l])
                maidens.append(temp[l + 1])
                no_balls.append(temp[l + 2])
                wide.append(temp[l + 3])

            temp = []
            for i in bowlers_table_2.find_all('div', class_="cb-col cb-col-10 text-right"):
                temp.append(i.text)
            del temp[0:2]
            for l in range(0, len(temp), 2):
                given_runs.append(temp[l])
                economy.append(temp[l + 1])
```

```python
bowling_data(href_values)
```
